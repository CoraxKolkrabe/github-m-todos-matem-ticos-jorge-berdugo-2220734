# Ejercicio 5 – Sección 1.1.6
*Definición (Ejemplo 1.2)*



*   $\vec r_{CM} = \frac{\sum_i m_i \vec r_i}{\sum_i m_i}$




```python
import sympy as sp

def centro_masa(masas, posiciones):
    # Verificación de dimensiones
    dim = posiciones[0].rows
    for r in posiciones:
        if r.rows != dim:
            raise ValueError("Todas las posiciones deben tener la misma dimensión")

    M = sum(masas)
    cero = sp.zeros(dim, 1)
    r_cm = cero

    for m, r in zip(masas, posiciones):
        r_cm += m * r

    return sp.simplify(r_cm / M)
```

(a) Masas 1,2,3,4 en un cuadrado de lado a = 2

```python
a = 2

masas = [1, 2, 3, 4]

posiciones = [
    sp.Matrix([0, 0]),
    sp.Matrix([a, 0]),
    sp.Matrix([a, a]),
    sp.Matrix([0, a])
]

centro_masa(masas, posiciones)
```

(b) Cubo de lado 2

*  Masas inferiores: 1,2,3,4

*  Masas superiores: 5,6,7,8

```python
a = 2

masas = [1,2,3,4,5,6,7,8]

posiciones = [
    sp.Matrix([0,0,0]),
    sp.Matrix([a,0,0]),
    sp.Matrix([a,a,0]),
    sp.Matrix([0,a,0]),
    sp.Matrix([0,0,a]),
    sp.Matrix([a,0,a]),
    sp.Matrix([a,a,a]),
    sp.Matrix([0,a,a])
]

centro_masa(masas, posiciones)

#habia un error que se debía a una incompatibilidad dimensional generada por la reutilización del entorno de ejecución.
#Se corrigió asegurando que el vector nulo inicial y todas las posiciones vivan en el mismo espacio vectorial.
```

```python
a + b + c + d
a + b - c - d
a - b + c - d
a - a + b - c + d
```

(b) Ángulo entre $a,b,c,d$

```python
def angulo(u, v):
    return sp.acos((u.dot(v)) / (sp.sqrt(u.dot(u)) * sp.sqrt(v.dot(v))))
```
